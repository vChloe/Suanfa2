```python
import numpy as np
```

```python
#随机生成顺序数组
def ShellSort(A):
    step = len(A) // 2
    while step > 0:
        for i in np.arange(step,len(A)):
            while i >= step and A[i] < A[i - step]:
                A[i],A[i - step] = A[i - step],A[i]
                i -= step
        step //= 2
    return A

Min = 0
Max = 1000
Num = 1000
A = np.random.randint(Min,Max,Num)
print('OriginalArray:\n',A,'\n')
B = ShellSort(A)
print('Sorted:\n',B,'\n')
```

```python
def SequentialSearch(A,k):
    N = len(A)
    for i in np.arange(N):
        if A[i] == k:
            index = i
            return index
    return -1

k = np.random.randint(Min,Max,1)
print(k)
index = SequentialSearch(B,k)
print('index:',index)
```

```python
def QuickSequentialSearch(A,k):
    N = len(A)
    for i in np.arange(N + 1):
        if k <= A[i]:
            if k == A[i]:
                return i
            else:
                return -1
            
k = np.random.randint(Min,Max,1)
print(k)
index = QuickSequentialSearch(B,k)
print('index:',index)
```

```python
def BinarySearch(A,k):
    N = len(A)
    l = 0
    u = N - 1
    index = -1
    while u != l:
        i = (l+u) // 2
        if k < A[i]:
            u = i-1
        if k > A[i]:
            l = i+1
        if k == A[i]:
            index = i
            break
        if u == l:
            break
    return index

k = np.random.randint(Min,Max,1)
print(k)
index = BinarySearch(B,k)
print('index:',index)
```

```python
def UniformBinarySearch(A,k):
    N = len(A)
    i = N//2 + 1
    m = N // 2
    index = -1
    while m != 0:

        if k < A[i]: 
            if m == 0:
                break
            i -= m//2 + 1
            m = m // 2
            if i < 0:
                i = 0
            if i > N-1:
                i = N-1
        if k > A[i]:
            if m == 0:
                break
            i += m//2 + 1
            m = m // 2
            if i < 0:
                i = 0
            if i > N-1:
                i = N-1
        if k == A[i]:
            index = i
            break
    return index

k = np.random.randint(Min,Max,1)
print(k)
index = UniformBinarySearch(B,k)
print('index:',index)
```
